```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/heart-disease-dataset.csv")
df.head()
```

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

df.info()
```

```python
df.isnull().sum()
```

# Histogram, Boxplot, Countplot, Correlation heatmap

```python
#  Histograms to see feature distributions
df.hist(figsize=(12,10))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="target", data=df)
plt.title("Target Class Distribution")
plt.xlabel("Heart Disease Presence")
plt.ylabel("Count")
plt.show()
```

```python
#  Correlation heatmap to explore feature relationships
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#  Boxplots to detect outliers in numeric features
plt.figure(figsize=(12,8))
df.boxplot()
plt.xticks(rotation=45)
plt.title("Boxplots of Numerical Features")
plt.show()
```
